```python
class Students:

    def __init__(self,math:int,physics:int,chemistry:int):
        self.math = math
        self.physics = physics
        self.chemistry = chemistry

    def average_marks(self):
        average = self.math + self.physics + self.chemistry
        average = average / 3

        return round(average, 2)
```

```python
s1 = Students(
    math=50,physics=70,chemistry=80
)
```

```python
s1.average_marks()
```

```python
class Students:

    def __init__(self,name:str,subjects_marks:dict):
        self.name = name
        self.subjects_marks = subjects_marks

    def average_marks(self):
        summ = 0
        total_subjects = 0
        for _ , val in self.subjects_marks.items():
            total_subjects+=1
            summ += val
        return f"{self.name}, Your avg marks are {summ/total_subjects}"
```

```python
subjects_marks = {
    'English':60,
    'Math':70,
    'Chemistry':80
}
```

```python
subjects_marks.items()
```

```python
for key, val in subjects_marks.items():
    print(key,val)
```

```python
s1 = Students(name='Mani',subjects_marks=subjects_marks)
```

```python
s1.average_marks()
```

## Static Mehtods

```python
class Life:

    def __init__(self):
        pass

    def hello():
        print("Hello Life")
```

```python
l1 = Life()
```

```python
class Life:

    def __init__(self):
        pass

    @staticmethod
    def hello():
        print("Hello Life")
```

```python
l1 = Life()
```

```python
l1.hello()
```
