```python
from pydantic import validate_call

class Student:

    @validate_call
    def __init__(self, name:str):
        self.name = name
        print(name)
```

```python
s1 = Student(name="1")
```

```python
from pydantic import validate_call, StrictStr

class Student:

    @validate_call
    def __init__(self, name:StrictStr):
        self.name = name
        print(name)
s1 = Student(name=1)
```

```python
class Jail:

    def __init__(self):
        print("jail")

    def __init__(self, crime:str):
        self.crime = crime

    def crime_name(self):
        print(self.crime)
```

```python
class Mobiles:

    mobile_type = 'smartphone'

    def __init__(self, brand:str):
        self.brand = brand
        print(self.brand)

    def ram(self, capacity:int):
        print(self.brand)
        print(self.mobile_type)
        print(capacity)

    def get_ram(self,capacity):
        print(self.ram(capacity))
```

```python
m1 = Mobiles(brand='samsung')
```

```python
m1.ram(capacity=8)
```

```python
m1.get_ram(capacity=9)
```
