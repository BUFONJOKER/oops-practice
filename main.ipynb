```python
class Students:

    def __init__(self,math:int,physics:int,chemistry:int):
        self.math = math
        self.physics = physics
        self.chemistry = chemistry

    def average_marks(self):
        average = self.math + self.physics + self.chemistry
        average = average / 3

        return round(average, 2)
```

```python
class Students:

    def __init__(self,name:str,subjects_marks:dict):
        self.name = name
        self.subjects_marks = subjects_marks

    def average_marks(self):
        summ = 0
        total_subjects = 0
        for _ , val in self.subjects_marks.items():
            total_subjects+=1
            summ += val
        return f"{self.name}, Your avg marks are {summ/total_subjects}"
```

```python
subjects_marks = {
    'English':60,
    'Math':70,
    'Chemistry':80
}
```

```python
subjects_marks.items()
```

```python
for key, val in subjects_marks.items():
    print(key,val)
```

```python
s1 = Students(name='Mani',subjects_marks=subjects_marks)
```

```python
s1.average_marks()
```

```python
from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    @abstractmethod
    def pay(self):
        pass

class CardPay(PaymentGateway):
    def pay(self):
        print("Paid using Card")

class UpiPay(PaymentGateway):
    def pay(self):
        print("Paid using UPI")

class Purchase:
    def __init__(self, gateway):
        self.gateway = gateway

    def checkout(self):
        print("Checkout started")
        self.gateway.pay()

p1 = Purchase(CardPay())
p2 = Purchase(UpiPay())

p1.checkout()
p2.checkout()
```

```python
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self,color:str,type:str):
        print(f"{self.color} color {self.type}'s engine is starting")

class Ford(Vehicle):
    def start_engine(self, color, type='car'):
        return super().start_engine(self.color,self.type)

class Ferrari(Vehicle):
    def start_engine(self, color, type='car'):
        return super().start_engine(self.color,self.type)

class Car:
    def __init__(self,brand,color:str):
        self.brand = brand
        self.color = color

    def start(self):
        return self.brand.start_engine(self.color)
```

```python
c1 = Car(brand=Ferrari(), color='black')
```

```python
c1.start()
```

```python
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self, color:str, type:str, brand:str):
        print(f"{brand} {color} color {type}'s engine is starting")

class Ford(Vehicle):
    def start_engine(self, color, type='car', brand='Ford'):
        return super().start_engine(color, type, brand)

class Ferrari(Vehicle):
    def start_engine(self, color, type='car', brand='Ferrari'):
        return super().start_engine(color, type, brand)

class Car:
    def __init__(self, brand: Vehicle, color:str):
        self.brand = brand
        self.color = color

    def start(self):
        return self.brand.start_engine(self.color)
```

```python
c1 = Car(brand=Ferrari(), color='black')
```

```python
c1.start()
```

```python
class Employee:
    def __init__(self, name, age):
        self.name = name       # public
        self._age = age        # protected

class SubEmployee(Employee):
    def show_age(self):
        print("Age:", self._age)   # Accessible in subclass

emp = SubEmployee("Ross", 30)
print(emp.name)        # Public accessible
emp.show_age()         # Protected accessed through subclass
```
